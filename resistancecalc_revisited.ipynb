## [Previously](http://nbviewer.ipython.org/github/joshualim/ipython/blob/master/20150226_resistancewirecalc/resistancecalc.ipynb), I addressed a misconception that effective heating elements should have a low resistance. Now, we extend this to think about how small should the resistance be?
Heat is generated in an electric circuit when current flows: the greater the current, the greater the heat output. For designing a heating element this tells us that we should pick something with a low resistance to allow lots of current to flow. The question is how small can we make the resistance of the element and have a good heating element: 10 Ohm, 1 Ohm, 0.001 Ohm?  

#### Back to the last problem
Recall the simple heating element circuit from the [previous example](http://nbviewer.ipython.org/github/joshualim/ipython/blob/master/20150226_resistancewirecalc/resistancecalc.ipynb), which is made of a resistor (with resistance, R), battery (supplying a voltage Vsupply) and connecting wires. 

<center> <img src="files/circuit1.png" /> </center>

By having a small resistance, we maximise the current and therefore the heating - for the simple circuit we used Ohm's law to find the current $I=\frac{V}{Relement}$. By this reasoning as we the resistance of the element go towards zero we will get infinite current out - clearly this is nonsense and we must be missing something from our model!

In solving this problem previously we ignored two things:
1. the wires have some small resistance ($R_{wire}$).
2. the battery has an internal resistance.

Why do I care about these now? Well, as we make the resistance of the element closer to zero, we are effectively ignoring this so we should take care of these other contributions.

## Why was ignoring the wire resistance and internal resistance ok before?
To answer this we have to revisit our simple circuit and start to think about the things we left out before. Our new circuit diagram looks like this, where we now also have $R_{wire}$ and Rint (for the wire and internal resistances).
<center> <img src="files/circuit3.png" /> </center>
#####1. the wires have some small resistance ($R_{wire}$).
Previously we ignored the resistance of the wire as the resistance of the element was much bigger, so the small wire resistance made no difference to the current. We can see this by rearranging Ohm's law for a series circuit (and including $R_{wire}$): 

$I=\frac{V_{supply}}{Rwire+Relement}$

for the case where $R_{element}$ >> $R_{wire}$, we will have :

$I=\frac{V_{supply}}{Relement}$

This is just Ohms law for the simple one resistor circuit we considered before, where the $R_{wire}=0$. In our new problem we are making the $R_{element}$ small, so our assumption that $R_{wire}$ is negligible should be revisited.


#####2. We also previously ignored the internal resistance of the cell.

We can argue that it was valid to ignore this again for similar reasons to before (typically the internal resistance is very small and measured in milli Ohms). Also to help us on the next part of the problem, we will also assume that we have a near perfect voltage source (i.e. Rinternal << $R_{wire}$). In this way we can make our model slightly more realistic without introducting too much additional complexity.

## New model for the circuit 
Our new model for the circuit is now shown and comprises of two resistors: one for the wire and one for the heating element. 
<center> <img src="files/circuit2.png" /> </center>
We can calculate the power dissipated across a resistor by $P=IV$. As we are interested in the power output of the heater we can apply this to our element and write:

$P_{element} = I V_{element} = I^2 R_{element}$

where the $P_{element}$ is the power distributed across the element and the second inequality we get by using Ohms law for the voltage across the element.

We can write a similar statement for the power used in the wire, $P_{wire}$:

$P_{wire} = I V_{wire} = I^2 R_{wire}$

You will notice the same current, $I$, is used in both equations: this is because any current that flows through the element must also flow through the wire and so this is the same current, $I$. By conservation of energy we can also define the total power in the circuit, $P_{tot}$:

$P_{tot} = P_{element} + P_{wire}$

If we now pick some values for the voltage supply, properties of the wire, etc.; we can go ahead and calculate what the power dissipated will look like as we change the resistance of the element. This is covered in the [next blog.](http://nbviewer.ipython.org/github/joshualim/ipython/blob/master/20150226_resistancewirecalc/resistancecalc_revisited_calc.ipynb)

Before we do any calculation, it's important to think about what we intuitively expect the answer to be.

## The answer: with no calculation
From the equation $P_{element} = I^2 R_{element}$, this tells us how to have the highest heater power output: we should ideally have the current as large as possible (so a small resistance - we knew this from [before](http://nbviewer.ipython.org/github/joshualim/ipython/blob/master/20150226_resistancewirecalc/resistancecalc.ipynb)) and also the resistance as large as possible. Clearly it is not possible to have a very small and very large resistance for the heating element at the same time so our real answer lies in between!

Let's think about a few cases and imagine changing the resistance of the element to try and find the situation with the highest heater power. 

##### When $R_{wire} << R_{element}$
If the resistance of the element is very large then we know that the power output will be low (this is what we reasoned in the [previous example](http://nbviewer.ipython.org/github/joshualim/ipython/blob/master/20150226_resistancewirecalc/resistancecalc.ipynb)). 

As we decrease $R_{element}$ (while $R_{wire} << R_{element}$), we again know from our previous example that the power output of the element will increase as the current increases.

We can also think about what happens to the wire, Where we have the case $R_{wire} << R_{element}$, the power output by the wire will also be very small (recall $P_{wire} = I^2 R_{wire}$ compared to $P_{element} = I^2 R_{element}$).

##### When $R_{wire} \approx R_{element}$
Now let's think about the case where $R_{wire} \approx R_{element}$: by looking at our new model of the circuit, we can see that if we swap the position of the $R_{wire}$ and $R_{element}$ the circuit will behave in the exact same way. Clearly for the case when the two resistances are equal, the power used by both most be the same too!

##### When $R_{wire} > R_{element}$
Now we have the reverse of the first problem we solved, as $R_{wire}$ is now bigger than $R_{element}$. 

Now let's look at the equations describing the power: $P_{wire} = I^2 R_{wire}$ and compare it  to $P_{element} = I^2 R_{element}$. We notice that if the $R_{element}$ is now smaller than $R_{wire}$, there is less power used in the element compared to the wire. Now for the case where $R_{element}=0$ the power from the element will be zero and all the power used up in the wires (this would make a very bad and dangerous heater indeed as the wires would get very hot!)

## Summary
Without doing any exact calculation we now know that the power output by the heating element is low when
* we have very large $R_{element}$  values (we knew this before).
* we have very low $R_{element}$ values (where $R_{element} < R_{wire}$). 

At some in between value for $R_{element}$, we know that this power output reaches a maximum. To have the maximum $R_{element}$ power output we need 
* a high as possible current (as $P_{element} \propto I^2$) therefore a low resistance 
* a high resistance (as $P_{element} \propto R $).

These two opposing constaints are balanced and give the maximum when $R_{element} = R_{wire}$. We can understand this is as the lowest possible value $R_{element}$ can be, because lowering $R_{element}$ below this does not increase the current much as $R_{wire}$ will start to dictate the current through the circuit. This means decreasing the $R_{element}$ no longer gives a proportional increase in current and power through the element. 

We've also shown that the assumptions you make about a problem (like our idea to ignore the resistance of the wire), have implicatons - while it works for a simple case, by asking another question, we found that we need to improve our model. As physicists we try to make our models as simple as possible. But often, we have to include more detail (like the wire resistance) to match what we observe in the world around us. This is usual completely usual and important part of science - by going between theory and what we know to be true, we can build up descriptions of the world around us.
